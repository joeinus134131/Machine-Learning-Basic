```python
# Jawaban no.5
import numpy as np

fifa.columns = [col.replace(" ", "") for col in fifa.columns]

fifa['pace'] = fifa['pace'].str.lstrip()
fifa['shooting'] = fifa['shooting'].str.lstrip()
fifa['passing'] = fifa['passing'].str.lstrip()
fifa['dribbling'] = fifa['dribbling'].str.lstrip()

fifa['pace'] = fifa['pace'].replace("", np.nan)
fifa['shooting'] = fifa['shooting'].replace("", np.nan)
fifa['passing'] = fifa['passing'].replace("", np.nan)
fifa['dribbling'] = fifa['dribbling'].replace("", np.nan)

fifa['pace'] = fifa['pace'].astype(float)
fifa['shooting'] = fifa['shooting'].astype(float)
fifa['passing'] = fifa['passing'].astype(float)
fifa['dribbling'] = fifa['dribbling'].astype(float)
```

```python
# Jawaban no 6
fifa.isna().sum()
```

```python
# Jawaban no.7
mean_pace = round(fifa["pace"].mean())
fifa["pace"] = fifa["pace"].fillna(mean_pace)

mean_shoot = round(fifa["shooting"].mean())
fifa["shooting"] = fifa["shooting"].fillna(mean_shoot)

mean_pass = round(fifa["passing"].mean())
fifa["passing"] = fifa["passing"].fillna(mean_pass)

mean_drib = round(fifa["dribbling"].mean())
fifa["dribbling"] = fifa["dribbling"].fillna(mean_drib)

mean_pos = round(fifa["positioning"].mean())
fifa["positioning"] = fifa["positioning"].fillna(mean_pos)
```

```python
# Jawaban no. 8
fifa.query("rating == rating.max()")
```

```python
# Jawaban no. 9
fifa.query("rating == rating.min()")
```
